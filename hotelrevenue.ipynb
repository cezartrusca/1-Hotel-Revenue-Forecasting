```python
data = pd.read_csv('hotel_bookings.csv')

# Display the name of each column in the DataFrame
print(data.columns)

# Display the data types of each column
print(data.dtypes)

# Display the first 5 and last 5 rows
print(data)

# Give a general overview of the distribution
# count, mean, min/max
print("Statistics:\n", data.describe())

# Count all data that is not null for each column
# Data type of each column, number of columns, memory usage, and the number of records
print(data.info())

# View number of records (rows) and columns
print(data.shape)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

x = data.iloc[:,1:32]
y = data.iloc[:,-1]

best_features = SelectKBest(score_func=chi2, k=3)
fit = best_features.fit(x, y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)

features_scores = pd.concat([dfcolumns, dfscores], axis=1)
features_scores.columns = ['Features', 'Score']
features_scores.sort_values(by = 'Score')
```
